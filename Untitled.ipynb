# Data Rescaling 

```python
df_new = df[['Age', 'Fare', 'Pclass']]
df_new
```

```python
#MinMaxScaler
array = df_new.values
data_scaler = preprocessing.MinMaxScaler(feature_range = (0,1))
data_rescaled = data_scaler.fit_transform(array)
data_rescaled
```

```python
#Normalization
from sklearn.preprocessing import Normalizer
#nom_array = df_new.values
Data_normalizer = Normalizer(norm = 'l1').fit(df_new)
Data_normalized   = Data_normalizer.transform(df_new)
Data_normalized
```

```python
#Binarization
from sklearn.preprocessing import Binarizer
bi_array = df_new[['Pclass']]
data_binarizer = Binarizer(threshold = 2).fit(bi_array)
data_binarized = data_binarizer.transform(bi_array)
data_binarized
```
